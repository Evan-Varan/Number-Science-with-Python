# Problem 3:

```python
t, n = sp.symbols('t n')
T0 = sp.pi
f0 = 1/T0
w0 = 2*sp.pi*f0
x_t_function = sp.Heaviside(t + T0/4) - sp.Heaviside(t - T0/4) # one period

a0 = (1 / T0) * sp.integrate(x_t_function, (t, 0, T0))
an = (2 / T0) * sp.integrate(x_t_function * sp.cos(n * w0 * t), (t, 0, T0))
bn = (2 / T0) * sp.integrate(x_t_function * sp.sin(n * w0 * t), (t, 0, T0))

terms = 50
x_t = a0 + sp.summation(an * sp.cos(n * w0 * t) + bn * sp.sin(n * w0 * t), (n, 1, terms))

xtt = sp.lambdify(t,x_t)

sp.plot(x_t_function,x_t, (t, 0, 3), ylabel='x(t)', title='Fourier Series Approximation of x(t)')
```

# Problem 4:

```python
import numpy as np
import matplotlib.pyplot as plt

t, w = sp.symbols('t w', real=True)
x_t = sp.Heaviside(t + 1/2) - sp.Heaviside(t - 1/2) #from -1/2 to 1/2
X_w = sp.fourier_transform(x_t, t, w)
X_w_func = sp.lambdify(w, X_w, 'numpy') #For graphing
print(f'X_w = {X_w}')

wa = np.linspace(-10,10,100)
Xwmag = abs(X_w_func(wa))
Xwphase = np.angle(X_w_func(wa))

#Magnitude
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(wa, Xwmag)
plt.title("Magnitude of X(w)")
plt.xlabel("w")
plt.ylabel("Magnitude of X(w)")

#Phase
plt.subplot(122)
plt.plot(wa,Xwphase)
plt.title("Phase of X(w)")
plt.xlabel("w")
plt.ylabel("Phase of X(w)")
plt.show()
```
